```r
food <- c(188,146,326,224,278)
cloth <- c(312,254,457,376,398)
goods <- c(56,87,120,67,98)
elect <- c(298,193,525,336,277)
store <- c("札幌","仙台","東京","名古屋","大阪")
item <- c("食品","衣類","雑貨","家電")
uriage <- rbind(food, cloth,goods,elect)

colnames(uriage) <- store
rownames(uriage) <- item
uriage
```

```r
par(family = "HiraKakuProN-W3")
```

```r
barplot(uriage, legend=rownames(uriage))
title(main="店舗別売上", xlab="地域", ylab="金額")
```

```r
pie(apply(uriage,1,sum))
pie(apply(uriage,2,sum))
```
